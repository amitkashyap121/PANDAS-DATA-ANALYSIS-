###Add column with city_name with help of Address###

```python
def get_city(add):
    return add.split(",")[1]

def get_state(adds):
    return adds.split(",")[2].split(" ")[1]

all_months["City"] = all_months["Purchase Address"].apply(lambda x: get_city(x) + " "+ get_state(x))   # important
all_months
```

```python
result=all_months.groupby("City").sum()
result
```

```python
x=[city for city, df in all_months.groupby("City")]

plt.bar(x, result["Sales"])
plt.title("Sales Graph")
#plt.legend()
plt.xticks(x ,rotation = "vertical")#size =12) 
plt.xlabel("City names")
plt.ylabel("Sales in doller")
plt.show()
```

```python
# what orders are most sold together ?
df =all_months[all_months["Order ID"].duplicated(keep=False)]
df
```

```python
df
```

```python
df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df2 = df[['Order ID', 'Grouped']].drop_duplicates()
df2
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in df2['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))

for key,value in count.most_common(10):
    print(key, value)
```
